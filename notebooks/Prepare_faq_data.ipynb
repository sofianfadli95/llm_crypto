```python
import pandas as pd
```

```python
# Open both text files
documents = []
with open('question.txt', 'r', encoding='utf-8', errors='ignore') as questions_file, open('answer.txt', 'r', encoding='utf-8', errors='ignore') as answers_file:
    
    # Step 2: Loop through both files simultaneously
    for question, answer in zip(questions_file, answers_file):
        print(f"Question: {question.strip()} | Answer: {answer.strip()}")
        documents.append({'question': question.strip(), 'answer': answer.strip()})
```

```python
import json
with open('faq_crypto.json', 'wt') as f_out:
    json.dump(documents, f_out, indent=2)
```
